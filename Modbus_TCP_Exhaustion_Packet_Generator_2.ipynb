```python
from scapy.all import *

import time
import random
import csv
import subprocess

import cProfile
import pstats

target_ip = "192.168.56.112"
target_port = 502
excluded_ip = target_ip
random_src_port = random.randint(1024, 65535)
prefixForIPgenerating = "192.168.56."
INTERFACE = "Ethernet 4"
csv_filename = "Logging_Data3.csv"

HandshakeSUCCESS = 0
HandshakeFAIL = 0
HandshakeSUM = 0

arp_counter = 0
captured_packet = None


def get_mac_address_via_ip(target_ip):
    try:
        result = subprocess.run(['arp', '-a', target_ip], capture_output=True, text=True, check=True)
        print(result.stdout)
        
         # Searching for the MAC address in standard format
        mac_address_match = re.search(r'([0-9A-Fa-f]{2}[:-]){5}([0-9A-Fa-f]{2})', result.stdout)

        if mac_address_match:
            mac_address = mac_address_match.group(0).replace("-", ":")  # Use colons instead of hyphens
            print(f"MAC Address: {mac_address}")
            return mac_address
        else:
            print("MAC Address not found in the output.")
            return False
    except subprocess.CalledProcessError as e:
        print(f"Error: {e}")
        print(f"Output: {e.output}")
        return False


def check_port(remote_host, port):
    try:
        # Create a socket object
        sock = socket.create_connection((remote_host, port), timeout=2)

        # If the connection is successful, the port is open
        print(f"Port {port} is open on {remote_host}")
        sock.close()
        return True

    except (socket.timeout, ConnectionRefusedError):
        # If the connection fails, the port is closed
        print(f"Port {port} is closed on {remote_host}")
        return False
    

def create_IP_number_list():
    numbers = list(range(231, 254 + 1))
    return [num for num in numbers if num not in  {112, 113, 161}]


def generate_random_ip(ip_list):
    if not ip_list:
        return None
    
    random_octet = random.choice(ip_list)
    ip_list.remove(random_octet)
    ip_address = f"192.168.56.{random_octet}"
    return ip_address


def generate_random_mac():
    mac = [random.choice("0123456789ABCDEF") for _ in range(12)]
    return ":".join(["".join(mac[i:i+2]) for i in range(0, 12, 2)])
    

class ModbusPDU(Packet):
    name = "Modbus PDU"
    fields_desc = [
        ShortEnumField("function_code", 3, {1: "Read Coils", 3: "Read Holding Registers"}),
        XShortField("address", 100),
        XShortField("count", 1)
    ]

class ModbusADU(Packet):
    name = "Modbus ADU"
    fields_desc = [
        XShortField("transaction_id", 1),
        XShortField("protocol_id", 0),
        ShortField("length", None),
        PacketField("pdu", None, ModbusPDU),
    ]

    def post_build(self, pkt, pay):
        if self.length is None:
            l = len(pay) + 2
            pkt = pkt[:4] + struct.pack("!H", l) + pkt[6:]
        return pkt + pay


def send_modbus_request(target_ip, random_source_ip, source_port=random_src_port, target_port=target_port):
    modbus_request = IP(src=random_source_ip, dst=target_ip) / \
                      TCP(sport=random_src_port, dport=target_port) / \
                      ModbusADU()
    send(modbus_request, iface=INTERFACE)


def print_modbus_response(response):
    if response:
        print(response)
    else:
        print("No response received")




def stop_sniffer(pkt):
    global captured_packet
    if pkt.haslayer(ARP) or (pkt.haslayer(TCP) and pkt[TCP].flags & 0x12):  # check SYN-ACK 
        captured_packet = pkt
        return True
    return False

def start_sniffer():
    global captured_packet
    captured_packet = None
    sniffer = AsyncSniffer(iface=INTERFACE, store=False, filter="arp or tcp", stop_filter=stop_sniffer)
    sniffer.start()
    return sniffer

def run_sniffer():
    sniffer = start_sniffer()
    sniffer.join()


def evaluate_packet(packet, target_ip):
    # Check if it's an ARP packet
    if packet.haslayer(ARP):
        # Check if the ARP packet is looking for the target
        if packet[ARP].op == 1:
            if packet[ARP].pdst == target_ip:
                print("ARP Who-Has Request for {}".format(target_ip))
                return"ARP Request"
            else:
                print("ARP Request - NONE")
                return"Unrelated ARP Request"        
        else:
            print("ARP response or other ARP packet:", packet.summary())
            return "ARP Response/Other"

    # Check if it's a TCP packet
    elif packet.haslayer(TCP)and packet.haslayer(TCP) and packet[TCP].flags & 0x12:
        # Check if we received a SYN-ACK response
        if packet[TCP].flags & 0x12:  
            print("TCP SYN-ACK response:", packet.summary())
            return "TCP SYN-ACK"
        else:
            print("Other TCP packet:", packet.summary())
            return "TCP Other"

    else:
        print("Other packet:", packet.summary())
        return "Other"
    
def send_fake_arp_response(random_source_ip, fake_mac):
    print("FAKE MAC sending")
    fake_arp_response = Ether(dst=target_MAC_addresse) / ARP(op=2, psrc=random_source_ip, hwsrc=fake_mac)
    sendp(fake_arp_response, iface=INTERFACE, verbose=False)
    print("FAKE MAC send")


target_MAC_addresse = get_mac_address_via_ip(target_ip)
isTheModbusPortEnableOnTarget = check_port(target_ip, target_port)
ip_list = create_IP_number_list()
time.sleep(1)

#<<<------>>> MAIN <<<------>>> MAIN <<<------>>> MAIN <<<------>>> MAIN <<<------>>> MAIN <<<------>>> MAIN <<<---
with open(csv_filename, mode='a', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    if csv_file.tell() == 0:
        csv_writer.writerow(["MAC", "IP", "Sent"])
    while True:
        fake_mac = generate_random_mac()        
        random_source_ip = generate_random_ip(ip_list)
        random_src_port = random.randint(1024, 65535)
        if not random_source_ip:
            print("TEST FINISHED")
            break
            
        ip_FAKE_ARP = random_source_ip
        syn_pkt = IP(src=random_source_ip, dst=target_ip) / TCP(sport=random_src_port, dport=target_port, flags="S", seq=0, window=64240, options=[('MSS', 1460), ('SAckOK', b''), ('WScale', 7), ('Timestamp', (int(time.time()), 0))])
        send_enable = True
        while True:
            HandshakeSUM += 1
            print("Random IP address:", random_source_ip)
            print("Random MAC address:", fake_mac)
            if send_enable:
                send(syn_pkt, verbose=False, iface=INTERFACE)
            run_sniffer()
            
            result = evaluate_packet(captured_packet, random_source_ip)
            if result == "ARP Request":
                send_fake_arp_response(random_source_ip,fake_mac)
                send_enable = False
            elif result=="TCP SYN-ACK":
                ack_pkt = IP(src=random_source_ip, dst=target_ip) / TCP(sport=random_src_port, 
                            dport=target_port, flags="A",
                            seq=captured_packet[TCP].ack,
                            ack=captured_packet[TCP].seq + 1)
                send(ack_pkt, iface=INTERFACE, verbose=False)
                HandshakeSUCCESS += 1
                print("TCP handshake SUCCESSFUL!")
                csv_writer.writerow([fake_mac, random_source_ip, "YES"])
                send_enable = True
                break
            else:
                print("An error occurred")
                send_enable = False
                HandshakeFAIL += 1
                print(result)

            print("SUM:",HandshakeSUM)
            print("SUCCES:",HandshakeSUCCESS)
            
        
print("summary SUM:",HandshakeSUM)
print("summary SUCCES:",HandshakeSUCCESS)
print("summary FAIL:",HandshakeFAIL)
```
